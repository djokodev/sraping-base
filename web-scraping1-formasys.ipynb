```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "http://feeds.bbci.co.uk/news/rss.xml"
```

```python
res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")
print(soup)
```

```python
items = soup.findAll("item")
items_list = []

for element in items:
    element_dict = {}
    
    element_dict['title'] = element.title.text
    element_dict['description'] = element.description.text
    element_dict['pubdate'] = element.pubdate.text
    
    items_list.append(element_dict)
    
items_list
    
    
```

```python
df = pd.DataFrame(items_list)
df
```

```python
df.to_csv("test1-web-scraping", index=False, encoding = 'UTF-8')
```
